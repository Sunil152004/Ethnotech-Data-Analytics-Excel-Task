```python
# Smart Hospital Bed Allocation System
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'patient_name': ["p1","p2","p3","p4","p5"],
    'age':[50,32,18,51,16],
    'condition':["normal","critical","low_serious","normal","critical"],
    'oxygen_level':[40,85,90,49,65]
})
print(df)
available_beds=4
print("available_beds= ",available_beds)
patients=df[(df['age']>50) | (df['condition'] == 'critical') | (df['oxygen_level']<50)]
print(patients)
for i in patients.index:
  patient_row=patients.loc[i]
  if available_beds>0:
    print(f"bed allocated for {patient_row['patient_name']}")
    available_beds-=1
  else:
    print(f"bed not allocated for {patient_row['patient_name']}")
# Bar Plot
plt.bar(df['patient_name'],df['age'])
plt.xlabel('patient_name')
plt.ylabel('age')
plt.title('age of patients')
plt.show()
print()
plt.bar(df['patient_name'],df['oxygen_level'])
plt.xlabel('patient_name')
plt.ylabel('oxygen_level')
plt.title('oxygen_level of patients')
plt.show()

#Pie Chart
plt.pie(df['age'],labels=df['patient_name'],autopct='%1.1f%%')
plt.title('age of patients')
plt.show()
print()
plt.pie(df['oxygen_level'],labels=df['patient_name'],autopct='%1.1f%%')
plt.title('oxygen_level of patients')
plt.show()
```
