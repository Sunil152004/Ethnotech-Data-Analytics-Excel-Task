Discriptive Analytics Using Python

```python
import pandas as pd

# Data to be saved to CSV
data = {
    'member_names': ["Sunil", "Gokul", "Sai", "Lakshmikanth", "Ganesh"],
    'ages': [22, 25, 24, 23, 26],
    'workout_hours': [2, 1.5, 2.5, 1, 3],
    'calories_burned': [500, 350, 600, 300, 750]
}

df_original = pd.DataFrame(data)

# Define the path where the CSV will be saved
csv_path = '/tmp/fitness_data.csv'

# Save the DataFrame to a CSV file
df_original.to_csv(csv_path, index=False)

print(f"Sample dataset saved to: {csv_path}")
print("Content of the generated CSV file:")
print(df_original.to_string(index=False))
```

```python
# Smart Fitness Center using Descriptive Analysis
import statistics as st
import matplotlib.pyplot as plt
member_names=["Sunil","Gokul","Sai","Lakshmikanth","Ganesh"]
ages=[22,25,24,23,26]
workout_hours=[2,1.5,2.5,1,3]
calories_burned=[500,350,600,300,750]
print("-----Fitness center Dataset------")
for i in range(len(member_names)):
  print(member_names[i],ages[i],workout_hours[i],calories_burned[i])
  # Mean Workout Hours
average_workout=sum(workout_hours)/len(workout_hours)
print("Average workout hours:",average_workout)
# Maximun Calories
max_calories=max(calories_burned)
print("Maximum calories burned:",max_calories)
# Minimum Calories
min_calories=min(calories_burned)
print("Minimum calories burned:",min_calories)
# Median Calories
median_calories=st.median(calories_burned)
print("Median calories burned:",median_calories)
# Standard Deviation
std_dev=st.stdev(calories_burned)
print("Standard deviation:",std_dev)
# Most Active Number
index=calories_burned.index(max_calories)
print("Most active member:",member_names[index])
# Bar Chart
plt.bar(member_names,workout_hours)
plt.xlabel("Member Names")
plt.ylabel("Calories Burned")
plt.title("Calories Burned By Members")
plt.show()
# Line Chart
plt.plot(member_names,calories_burned)
plt.xlabel("Member Names")
plt.ylabel("Calories Burned")
plt.title("Calories Burned By Members")
plt.show()
# Pie Chart
plt.pie(calories_burned,labels=member_names)
plt.title("Calories Burned By Members")
plt.show()
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)
print(df)
```

```python
# Filling missing values
df=df.fillna(0)
# convert Datatype
df['reading score']=df['reading score'].astype(int)
# Save Cleaned data
df.to_csv('Cleaned_data.csv',index=False)
print(df)
print("Data Cleaned Successfully")
```

```python
import matplotlib.pyplot as plt

# Bar chart
plt.figsize=(1,2)
plt.bar(df['gender'],df['math score'])
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.title('Math Score by Gender')
plt.show()

# Pie Chart
gender_counts = df['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Gender')
plt.show()

# Histogram Plot
plt.hist(df['math score'], bins=10, edgecolor='black')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Distribution of Math Scores')
plt.show()
```
